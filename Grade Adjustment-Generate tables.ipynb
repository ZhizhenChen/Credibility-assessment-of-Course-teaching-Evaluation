```python
import pandas as pd
import numpy as np

df = pd.read_excel("D:\\OneDrive - CUHK-Shenzhen\\Research\\Student profiling data\\Student profiling data\\1810-2020六学期学生成绩数据.xlsx")
group = df.groupby('ID')
k = []
v = []
for key, value in group:
    k.append(key)
    v.append(value)

tt = pd.read_excel("C:\\Users\\HUAWEI\\Desktop\\不在书院时间.xlsx")
```

```python
# Search for the same id in two tables
time_id = list(tt['id'])
agg = []
for i in k:
    if i in time_id:
        agg.append(i)
len(agg)
```

```python
# 生成包含时间id的成绩表格
# Generate table which includes time, id and grade
new_df = df.loc[df['ID'].isin(agg),:]
new_df = new_df.reset_index(drop = True)
new_df = new_df.drop(columns = ['Course Offering School','Term','Course ID'])
new_df
# new_df.to_excel('包含的成绩.xlsx')
```

```python
# 生成新的成绩表格（取平均）
# Generate new table with average grade
gg = new_df.groupby('ID')
update_df = pd.DataFrame()
for k,v in gg:
    if len(v) > 1:
        m = v['Grade'].mean()
        v=v.drop(index=v.index)
        v.loc[0] = [k, m]
    update_df = update_df.append(v)
update_df = update_df.reset_index(drop = True)
```

```python
# 生成包含时间的表格
# Generate new table with time included
new_tt = tt.loc[tt['id'].isin(agg),:]
new_tt = new_tt.reset_index(drop = True)
```

```python
# 合并两表
# concatenate two tables
new_time_grade = pd.concat([update_df,new_tt],axis = 1)
```

```python
# 替换
# Dropna
new_time_grade = new_time_grade.replace(86400,'NaN')
new_time_grade
```

```python
# 纵向删除空值
# Drop empty entries in columns
n = []

for i in new_time_grade.columns:
    value = 0
    for j in new_time_grade[i]:
        if str(j).isalpha():
            value += 1
    n.append(value)

idx = []
k = 0
for i in n:
    if i > 150:
        new_time_grade = new_time_grade.drop(columns = new_time_grade.columns[n.index(i)-k])
        k += 1
# new_time_grade.to_excel('相同成绩和时间.xlsx')
```
