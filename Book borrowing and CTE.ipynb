```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from pylab import *
```

```python
# 构造有成绩且有借阅的在1910和1810的id
# Collect id that has course grade and has borrowed books in fall 2018 and fall 2019 semesters.
df_g = pd.read_excel("C:\\Users\\HUAWEI\\Desktop\\有借阅的成绩id.xlsx")
df_b = pd.read_excel("C:\\Users\\HUAWEI\\Desktop\\有成绩的书籍.xlsx")

df_g['Tick'] = 0

group = df_b.groupby('Primary Identifier')
k = []
v = []
for key, value in group:
    k.append(key)
    v.append(value)
    
for i in range(len(df_g)):
    t = str(df_g['Term'].iloc[i])
    for j in range(len(k)):
        if k[j] == df_g['ID'].iloc[i]:
            if t == '1810':
                for m in v[j]['Loan Date']:
                    if m > pd.Timestamp('2018-09-01') and m < pd.Timestamp('2019-01-01'):
                        df_g['Tick'].iloc[i] = 1
                        break
                    else:
                        continue
            elif t == '1910':
                for m in v[j]['Loan Date']:
                    if m > pd.Timestamp('2019-09-01') and m < pd.Timestamp('2020-01-01'):
                        df_g['Tick'].iloc[i] = 1
                        break
                    else:
                        continue
            break
df_g
```

```python
df_g.to_excel('有成绩和借阅的id.xlsx')
```

```python
# 计算1810和1910的借阅比例
# Calculate the percentage of students who borrowed book in a specific course in 2018 fall and 2019 fall semesters.

four = []
# third =[]
second = []

group = df_g.groupby('Course ID')
k = []
v = []
for key, value in group:
    k.append(key)
    v.append(value)
    
for i in range(len(k)):
    f = v[i].apply(lambda x: v[i].Tick.loc[v[i]['Term'] == 1810].sum()/15)
#     f = v[i].apply(lambda x: v[2].Tick.loc[v[i]['Term'] == 1810].sum())
    s = v[i].apply(lambda x: v[i].Tick.loc[v[i]['Term'] == 1910].sum()/15)
    four.append(f[0])
#     third.append(t)
    second.append(s[0])
four
```

```python
# 生成表格
# Generate table
total_sum = pd.DataFrame({'Course':k,'1810':four,'1910':second})

total_sum
```

```python
# 合并成绩和借阅比例表格
# Concatenate tables with grade and borrow ratio.
grade = pd.read_excel("D:\\OneDrive - CUHK-Shenzhen\\Research\\Student profiling data\\Student profiling data\\1810-1820-1910三学期CTE_教务处.xlsx")

grade = pd.concat([grade,total_sum],axis = 1)

grade.columns = ['ID','Course18','Course19','Prof18','Prof19','new','1810','1910']
grade = grade.drop(columns = 'new')
grade
```

```python
# 作差准备画图
grade['delta_C'] = grade['Course19']-grade['Course18']
grade['delta_B'] = grade['1910']-grade['1810']
grade = grade.sort_values(by = 'delta_C')
grade = grade.reset_index(drop = True)
grade
```

```python
plt.scatter(grade['delta_C'],grade['delta_B'])
plt.title('CTE分数变化和借阅比例变化')
plt.xlabel('CTE成绩变化')
plt.ylabel('借阅比例变化')
mpl.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus']=False
plt.savefig("C:\\Users\\HUAWEI\\Desktop\\CTE分数变化和借阅比例.jpg")
plt.show()
```
