### 可视化不在书院时间
### Visualize time away from the college

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel("C:\\Users\\HUAWEI\\Desktop\\相同成绩和时间.xlsx")

# x = df.columns[2:df.shape[1]-1]
x = range(df.shape[1]-3)
y = df.iloc[-1,2:df.shape[1]-1]
fig = plt.figure(figsize = (20,10))

plt.xlabel('Date')
plt.ylabel('Time_Not_In_College')
plt.title('The distribution of time not in college')
# fig = plt.figure(figsize = (20,10))
plt.plot(x,y)
plt.savefig("C:\\Users\\HUAWEI\\Desktop\\不在书院时间分布图.jpg")
```
